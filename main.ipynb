```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Ajustar tamaÃ±o de la figura
plt.figure(figsize=(10, 5))

# Graficar la temperatura
plt.plot(df.loc['2017-01-01':'2017-01-02', 'Temperature'], linestyle='-', marker='o', color='b', label='Temperature')

# Formatear el eje X
ax = plt.gca()  # Obtener el eje actual

# Establecer el formato de fecha en el eje X
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))  # Formato de etiqueta

# **AquÃ­ se ajusta la separaciÃ³n entre ticks**
ax.xaxis.set_major_locator(mdates.HourLocator(interval=1))  # Coloca un tick cada 3 horas

# Rotar y alinear las etiquetas del eje X
plt.xticks(rotation=45, ha="right")

# Agregar tÃ­tulos y etiquetas
plt.title("EvoluciÃ³n de la Temperatura (1-2 Ene 2017)", fontsize=14)
plt.xlabel("Fecha y Hora", fontsize=12)
plt.ylabel("Temperatura (Â°C)", fontsize=12)

# CuadrÃ­cula para mejor lectura
plt.grid(True, linestyle='--', alpha=0.6)

# Agregar leyenda
plt.legend()

# Mostrar grÃ¡fico
plt.show()
```
